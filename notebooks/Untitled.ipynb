```python
import pandas as pd
from pycaret import classification
```

```python
data = pd.read_csv("../data/train.csv")
```

```python
data.head()
```

```python
classification_setup = classification.setup(data=data,target='Survived')
```

```python
classification.compare_models()
```

```python
classification_lr = classification.create_model('lr')
```

```python
classification.plot_model(classification_lr)
```

```python
classification.plot_model(classification_lr, plot = 'error')
```

```python
classification.plot_model(classification_lr, plot='feature')
```

```python
classification.evaluate_model(classification_lr)
```

```python
test_data = pd.read_csv("../data/test.csv")
```

```python
predictions = classification.predict_model(classification_lr, data=test_data)
```
